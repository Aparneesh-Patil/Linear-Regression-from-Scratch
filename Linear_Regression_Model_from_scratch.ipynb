```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('train.csv')
```

```python
def gradient_descent(m_now, b_now, L, points):
    m_gradient = 0.0
    b_gradient = 0.0

    n = len(points)
    for i in range(n):
      x = points.iloc[i].x
      y = points.iloc[i].y
      m_gradient += -(1/n) * x * (y - (m_now * x + b_now))
      b_gradient += -(1/n) * (y - (m_now * x + b_now))


    m = m_now - m_gradient * L
    b = b_now - b_gradient * L
    return m, b

m = 0.0
b = 0.0
L = 0.0001
for i in range(300):
  m, b = gradient_descent(m, b, L, data)

print(m, b)
```

```python
plt.scatter(data.x, data.y)
plt.plot(list(range(0, 100)), [m * x + b for x in range(0,100)], color='red')
plt.show()
```
